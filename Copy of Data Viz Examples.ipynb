```python
import matplotlib.pyplot as plt

# Sample data
category_amount = {
    'Category A qatar': 2,
    'Category B air india': 4 ,
    'Category C emirates':3 ,
    'Category D air cabada': 4,
    'Category E british airways': 2
}

# Extract category names and corresponding amounts
categories = list(category_amount.keys())
amounts = list(category_amount.values())

# Plotting the bar graph
plt.bar(categories, amounts)

# Customize the plot
plt.xlabel('Categories')
plt.ylabel('Amount')
plt.title('Bar Graph')
plt.xticks(rotation=45)

# Show the plot
plt.show()
```
